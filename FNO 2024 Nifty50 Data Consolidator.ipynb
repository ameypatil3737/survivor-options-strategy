```python
import pandas as pd
import glob
import os

# Mac folder path
options_folder = "/Users/nathdwar/Downloads/2024"

# find all csv files in folder and subfolders
files = glob.glob(os.path.join(options_folder, "**", "*.csv"), recursive=True)

print("Total CSV files found:", len(files))

all_data = []

for f in files:
    try:
        df = pd.read_csv(f)

        # standardize column names
        df.columns = [c.lower().strip() for c in df.columns]

        all_data.append(df)

        print("Loaded:", os.path.basename(f))

    except Exception as e:
        print("Error reading:", f, e)

# combine everything
options_df = pd.concat(all_data, ignore_index=True)

print("Rows after merge:", len(options_df))

# convert datetime column if present
if "datetime" in options_df.columns:
    options_df["datetime"] = pd.to_datetime(options_df["datetime"], errors="coerce")

if "date" in options_df.columns:
    options_df["date"] = pd.to_datetime(options_df["date"], errors="coerce")

# sort dataset
if "datetime" in options_df.columns:
    options_df = options_df.sort_values("datetime")

# remove duplicates
options_df = options_df.drop_duplicates()

options_df.reset_index(drop=True, inplace=True)

print("Final rows:", len(options_df))
```

```python
#Step 3 — (Recommended) Save as Parquet
#Parquet is much faster for backtesting.
```

```python
options_df.to_parquet("/Users/nathdwar/Python/Algo Trading/NIFTY_OPTIONS_2024_CONSOLIDATED.parquet")

print("Saved parquet version")
```
